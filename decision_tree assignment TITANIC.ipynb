```python
#decision tree assignment
import pandas as pd
df = pd.read_csv(r"C:\Users\user\Desktop\Adhil\titanic.csv")
df.head()
```

```python
#one hot encoding
dummies = pd.get_dummies(df.sex)
dummies.head()
```

```python
df2 = pd.concat([df,dummies],axis = 'columns')
df2.head()
```

```python
df1 =df2.drop(['passenger_id','sex','name','sib_sp','parch','ticket','cabin','embarked','fare'],axis = "columns")
df1.head()
```

```python
inputs = df1.drop('survived',axis = 'columns')
inputs.head()
```

```python
inputs.age[:]
```

```python
#data cleaning
inputs.age = inputs.age.fillna(inputs.age.mean())
inputs
```

```python
target = df.survived
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs,target,test_size = 0.003)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
